```python
resume = """
Nitesh Kumar
Skills: Python, SQL, Machine Learning, Data Analysis, Power BI
Experience: Java Intern at Prama Innovations
Education: B.E in Data Science
"""
```

```python
skills = ["Python", "SQL", "Machine Learning", "Data Analysis", "Power BI", "Java"]

found_skills = []

for skill in skills:
    if skill.lower() in resume.lower():
        found_skills.append(skill)

print("Skills Found:", found_skills)
```

```python
score = len(found_skills) * 15

print("Resume Score:", score, "%")
```
