```python
import pandas as pd

data = {
    "Product": ["Laptop","Phone","Tablet","Watch","Headphones"],
    "Sales": [500,700,300,200,400]
}

df = pd.DataFrame(data)

df
```

```python
question = input("Ask a question about sales: ")

if "highest" in question.lower():
    max_product = df.loc[df['Sales'].idxmax()]
    print("Highest sales product:", max_product["Product"])

elif "lowest" in question.lower():
    min_product = df.loc[df['Sales'].idxmin()]
    print("Lowest sales product:", min_product["Product"])

else:
    print("Question not recognized.")
```
