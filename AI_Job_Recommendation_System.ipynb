```python
resume_skills = ["Python", "Machine Learning", "SQL"]
```

```python
jobs = {
    "Data Scientist": ["Python", "Machine Learning", "Statistics"],
    "Data Analyst": ["SQL", "Excel", "Power BI"],
    "Machine Learning Engineer": ["Python", "Machine Learning", "Deep Learning"],
    "Software Developer": ["Java", "C++", "Algorithms"]
}
```

```python
recommended_jobs = []

for job, skills in jobs.items():
    match = set(resume_skills) & set(skills)

    if len(match) > 0:
        recommended_jobs.append(job)

print("Recommended Jobs:")
for job in recommended_jobs:
    print(job)
```
